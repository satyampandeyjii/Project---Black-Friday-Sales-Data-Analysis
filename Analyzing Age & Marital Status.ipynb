```python
import pandas as pd
```

```python
df = pd.read_csv('BlackFriday.csv')

del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
df.groupby('Age').size().plot(kind = 'bar', figsize = (10, 4), title = 'Purchase Distribution by Age')
```

```python
lst = []
for i in df['Age'].unique():
    lst.append([i, df[df['Age'] == i]['Product_ID'].nunique()])
    
data = pd.DataFrame(lst , columns = ['Age','Products'])
```

```python
data.plot.bar(x = 'Age', figsize = (8,6))
```

```python
df.groupby('Age').sum()['Purchase'].plot(kind = 'bar', figsize = (10, 4), title = 'Amount Spend by Age')
```

```python
df.groupby('Marital_Status').size().plot(kind = 'pie', autopct = '%0.1f')
```

```python
df.groupby('Gender').size().plot(kind = 'pie', autopct = '%0.1f')
```
